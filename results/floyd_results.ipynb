```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("floyd.csv")
df_columns = df.keys()
df.head()
```

```python
# Overview of Floyd
df.plot(x=df_columns[0], y=df_columns[1:], kind="line", marker="o")
plt.xticks(rotation=25)
plt.xlabel('Number of nodes')
plt.ylabel('Time Taken')
plt.title('Floyd\'s Algorithm - Time taken for different number of threads', fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# Serial vs Parallel Floyd - 2 Threads
df.plot(x=df_columns[0], y=[df_columns[1], df_columns[2]], kind="line", marker="o")
plt.xticks(rotation=25)
plt.xlabel('Number of nodes (N)')
plt.ylabel('Time Taken (us)')
plt.title('Floyd\'s Algorithm - Serial vs Parallel Execution Time', fontsize=20)
plt.grid()
plt.legend()
plt.show()
```
